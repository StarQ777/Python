```python
additional_data=[
    {'name':'ram','age':17,'branch':'lalit','Salary':34},
    {'name':'syan','age':18,'branch':'ktm','Salary':36},
    {'name':'arjun','age':19,'branch':'but','Salary':38},
    {'name':'bhim','age':20,'branch':'chit','Salary':40},
    {'name':'udistir','age':22,'branch':'pok','Salary':44},
    {'name':'laxman','age':24,'branch':'chit','Salary':48},
    {'name':'sakuni','age':23,'branch':'but','Salary':46},
    {'name':'kunti','age':19,'branch':'ktm','Salary':38},
    {'name':'karan','age':18,'branch':'lalit','Salary':36},
    {'name':'sita','age':20,'branch':'pok','Salary':40},
    {'name':'drapati','age':28,'branch':'birat','Salary':52},

]

new_records=pd.DataFrame(additional_data)
df=pd.concat([df,new_records],ignore_index=True)

df
```

```python
df=df.rename(columns={'Salary':'salary'})
df['salary']=df['salary'].astype(str)+'k'
df
```

```python
new_data=[
    {'name':'nakul','age':'','branch':'pal','salary':50},
]
df=pd.concat([df,pd.DataFrame(new_data)],ignore_index=True)
```

```python
#value Counts
print("\nBranch Distribution:")
print(df['branch'].value_counts())
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(15,5))

#1 PIe Chart branch distribution 
plt.subplot(131)
df['branch'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distirbution of Branch')

#2 Bar chart salary by sity 
plt.subplot(132)
avg_salary=df.groupby('branch')['salary'].apply(lambda x:x.str.replace('k','').astype(float).mean())
avg_salary.plot(kind='bar')
plt.title('Average salary by city')
plt.xticks(rotation=45)
plt.ylabel('Average salary (k)')

#3. scatter plot age vs salary 
plt.subplot(133)
salary_numeric=df['salary'].str.replace('k','').astype(float)
plt.scatter(df['age'],salary_numeric)
plt.xlabel('age')
plt.ylabel('salary(k)')
plt.title('age vs salary')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt 
x=[0,1,2,3,4,5,6]
y=[3,5,6,9,6,5,8,]

plt.figure(figsize=(6,4),dpi=80)

plt.plot(x,y,'r.-',label="2x")

x2=np.arange(0,4.5,0.5)

plt.plot(x2[4:],x2[4:]**3,'b*--',label='x^3')
plt.plot(x2[0:5],x2[0:5]**3,'g.-',label='x^3')

plt.title("My first graph!",fontsize=20)
plt.rcParams["font.family"]="monospace"
plt.rcParams["font.monospace"]=["FreeMono"]

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([0,1,2,3,4,5,6])

plt.legend()

plt.savefig('mygraph.png',dpi=200)

plt.show()
```
