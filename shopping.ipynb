```python
# Define the available items in different sectors with their prices
sectors = {
    'clothes': {
        '1. T-shirt': 20,
        '2. Jeans': 40,
        '3. Jacket': 60,
        '4. Shoes': 50
    },
    'beauty_products': {
        '1. Lipstick': 15,
        '2. Foundation': 25,
        '3. Perfume': 30,
        '4. Nail Polish': 10
    },
    'electronics': {
        '1. Headphones': 80,
        '2. Smartphone': 300,
        '3. Laptop': 700,
        '4. Smartwatch': 150
    },
    'groceries': {
        '1. Milk': 2,
        '2. Bread': 1,
        '3. Eggs': 3,
        '4. Butter': 4
    }
}

# Function to display items in a sector
def display_items(sector):
    print(f"\nItems available in {sector}:")
    for item, price in sectors[sector].items():
        print(f"{item}: ${price}")

# Function to handle shopping
def shopping():
    total_money = float(input("Enter the total amount of money you have: "))
    cart = []

    while total_money > 0:
        print("\nSectors available:")
        print("1. Clothes")
        print("2. Beauty Products")
        print("3. Electronics")
        print("4. Groceries")
        print("5. Exit")
        
        sector_choice = input("Enter the number of the sector you want to shop from: ")

        if sector_choice == '5':
            break

        sector_map = {
            '1': 'clothes',
            '2': 'beauty_products',
            '3': 'electronics',
            '4': 'groceries'
        }

        if sector_choice in sector_map:
            sector = sector_map[sector_choice]
            display_items(sector)
            item_choice = input("Enter the number of the item you want to add to the cart: ")

            item_map = {item.split('. ')[0]: item for item in sectors[sector].keys()}

            if item_choice in item_map:
                item = item_map[item_choice]
                item_price = sectors[sector][item]
                quantity = int(input(f"Enter the quantity of {item} you want to add: "))
                total_cost = item_price * quantity
                if total_money >= total_cost:
                    cart.extend([item] * quantity)
                    total_money -= total_cost
                    print(f"Added {quantity} x {item} to the cart. Remaining money: ${total_money:.2f}")
                else:
                    print("You don't have enough money for this item.")
            else:
                print("Item not found in this sector.")
        else:
            print("Sector not found.")

    print("\nShopping finished.")
    print("Items in your cart:", cart)
    print(f"Remaining money: ${total_money:.2f}")

# Start the shopping experience
shopping()
```
