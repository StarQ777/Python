##### Q8. Write a code that demonstrates the use of floating point numbers and arithmetic operations.

```python

num1 = 5.75
num2 = 2.5
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
modulus = num1 % num2
print("Addition:", addition)          
print("Subtraction:", subtraction)   
print("Multiplication:", multiplication)
print("Division:", division)           
print("Modulus:", modulus)
```

#### Q9. Write a code to demonstrate the difference between mutable and immutable types.

```python
# immutable
immutable_str = "Hello"
print("Original string:", immutable_str)

# Attempt to change the string
try:
    immutable_str[0] = "h"
except TypeError as e:
    print("Error:", e)

# Mutable
mutable_list = [1, 2, 3]
print("Original list:", mutable_list)

# Change the list
mutable_list[0] = 10
print("Modified list:", mutable_list)
```
